```python
import pandas as pd
import matplotlib.pylab as plt
plt.rcParams['figure.dpi'] = 200

col_names = ["submission_id", "session","date","level","code","server_error","client_error","version", "lang", "email", "username", "is_test", "demo", "start"]
df = pd.read_csv("../../Data/2-logs-plain-2.csv", names=col_names)
df = df.iloc[1:]

#Catch the string, convert the levels, then count the levels with server errors
df['level'] = pd.to_numeric(df.level, errors='coerce')
df = df.loc[df['level'] < 14]

print(df.shape[0])
print(df['client_error'].value_counts())
```

```python
#Let's extract some general information
total_amount = df.shape[0]
total_sessions = len(df['session'].unique())

print("Total amount of submissions: " + str(total_amount))
print("Total amount of sessions: " + str(total_sessions))
```

```python
df_client_error = df.loc[df['client_error'] != "-"]
total_client_error = df_client_error.shape[0]
client_error_percentage = round(total_client_error / total_amount * 100, 3)
print("Total amount of client errors: " + str(total_client_error))
print("Percentage of client errrors: " + str(client_error_percentage) + "%")
```

```python
df_server_error = df.loc[df['server_error'] != "None"]
df_server_error = df_server_error.loc[df_server_error['server_error'] != "-"]

total_server_error = df_server_error.shape[0]
server_error_percentage = round(total_server_error / total_amount * 100, 3)
print("Total amount of server errors: " + str(total_server_error))
print("Percentage of server errrors: " + str(server_error_percentage) + "%")
```

```python
unique_server_errors = len(df_server_error['server_error'].unique())
print("Total amount of unique server errors: " + str(unique_server_errors))
```

```python
#Here we filter all errors that only occur once
v = df_server_error['server_error'].value_counts()
df_multiple_server_error = df_server_error[df_server_error['server_error'].isin(v.index[v.gt(2)])]
amount_multiple_server_errors = len(df_multiple_server_error['server_error'].unique())
print("Server errors that occur multiple times: " + str(amount_multiple_server_errors))
```

```python
server_error_counts = df_multiple_server_error['server_error'].value_counts()
for index, value in server_error_counts.items():
    print(str(value) + ": " + str(index))
    print("--------------------------------------------------")
```

```python
print_errors = 0
for index, value in server_error_counts.items():
    if "print is" in index and "level 3" in index:
        print(str(value) + ": " + str(index))
        print("--------------------------------------------------")
        print_errors += value
print("Total amount of print errors in level 3: " + str(print_errors))
```

```python
if_errors = 0
for index, value in server_error_counts.items():
    if "if is" in index and "level 3" in index:
        print(str(value) + ": " + str(index))
        print("--------------------------------------------------")
        if_errors += value
print("Total amount of if errors in level 3: " + str(if_errors))
```

```python
wrong_level_errors = 0
for index, value in server_error_counts.items():
    if "goede code" in index or "right code" in index:
        print(str(value) + ": " + str(index))
        print("--------------------------------------------------")
        wrong_level_errors += value
print("Total amount of wrong level errors " + str(wrong_level_errors))
```

```python
wrong_keyword_errors = 0
for index, value in server_error_counts.items():
    if "is not a Hedy level" in index or "is geen commando in Hedy" in index:
        print(str(value) + ": " + str(index))
        print("--------------------------------------------------")
        wrong_keyword_errors += value
print("Total amount of wrong keyword errors " + str(wrong_keyword_errors))
```
